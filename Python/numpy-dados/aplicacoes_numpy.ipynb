###  Equao da reta

- Funo: y=ax+b


Imagine a equao da reta como uma receita para desenhar uma linha reta em um grfico. 
y:   como o resultado final da receita, a altura da linha em um determinado ponto.
x:  como o ingrediente principal, a posio horizontal no grfico.
a:  como a quantidade de ingrediente que voc usa para cada "poro" de x. Se "a" for grande, a linha ser inclinada para cima. Se "a" for pequeno, a linha ser mais horizontal.
b:  como a quantidade de ingrediente que voc coloca na receita antes de comear a usar o ingrediente "x".  o ponto onde a linha cruza o eixo vertical (y) do grfico.

Ento, a equao y=ax+b  como uma receita que te permite desenhar uma linha reta no grfico, ajustando a inclinao (a) e o ponto de partida (b) da linha. Voc pode pensar em exemplos do dia a dia:
- Preo de um txi: 
Imagine que o preo de um txi  R$ 5,00 de bandeirada (b) mais R$ 2,00 por quilmetro rodado (a). A equao seria y = 2x + 5, onde y  o preo total, x  a distncia percorrida, a  o preo por quilmetro e b  a bandeirada.
-Custo de produo:
Imagine que o custo de produo de um produto  R$ 10,00 de custo fixo (b) mais R$ 5,00 por unidade produzida (a). A equao seria y = 5x + 10, onde y  o custo total, x  a quantidade produzida, a  o custo por unidade e b  o custo fixo.

Espero que isso tenha ajudado a entender melhor a equao da reta!

O coeficiente angular, representado por "a",  calculado atravs de uma frmula que envolve a multiplicao dos arrays X e Y, suas somas e o nmero de elementos (n). A frmula :

python
Copiar cdigo
'''
a = (n*np.sum(X*Y)  np.sum(X)*np.sum(Y))/(n*np.sum(X**2)  np.sum(X)**2)
'''
Essa frmula nos permite encontrar a inclinao da reta que melhor se ajusta aos dados dos arrays X e Y.

Em outras palavras, o coeficiente angular indica o quanto a reta sobe ou desce para cada unidade que avanamos no eixo X.

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = (n*np.sum(X*Y)- np.sum(X)*np.sum(Y))/(n*np.sum(X**2)- np.sum(X)**2)
```

Coeficiente Linear
b = np.mean(Y)- a*np.mean(X)             
Ela  calculada utilizando a mdia dos valores do array Y, o coeficiente angular "a" e a mdia dos valores do array X.O coeficiente linear representa o ponto em que a reta intersecta o eixo Y. Lembrese que o coeficiente angular "a" j foi calculado anteriormente, ento voc pode usar o valor que encontrou para calcular o coeficiente linear "b". 

```python
b = np.mean(Y) -a*np.mean(X) 
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y)
```

 A funo random.seed()  como um "chaveiro" para a gerao de nmeros aleatrios. Ela define um ponto de partida para o gerador de nmeros aleatrios, garantindo que a mesma sequncia de nmeros seja gerada sempre que voc usar a mesma "chave" (semente). Imagine que voc tem um jogo de dados. Se voc sempre comear jogando os dados da mesma forma (por exemplo, segurandoos na mo direita e jogando para cima), voc sempre ter o mesmo resultado. A random.seed() funciona da mesma forma, definindo um ponto de partida para o gerador de nmeros aleatrios. Isso  muito til para garantir a reprodutibilidade dos seus resultados, ou seja, para que outras pessoas possam obter os mesmos resultados que voc.
