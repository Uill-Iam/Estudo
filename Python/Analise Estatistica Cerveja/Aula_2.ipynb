```python
# vendas  = unidades
# pub = Reais
# comercial = minutos

#Desenvolver um modelo de Regressão.

#A Altura dos Filhos quando os pais são altos, os filhos são baixos.  <-- exemplo

#1 - variavel publicidade e comercial consegue aumentar minhas vendas?  e qual delas é melhor  ? 


#modelos lineares
sns.lmplot(x = 'Publicidade' , y = "Vendas", data=vendas)

##y = a + (x1*b1)
```

```python
sns.lmplot(x = 'Comercial' , y = "Vendas", data=vendas)
```

```python
# r2 verifica a variabilidade que X explica Y
# 0 -- 1 
#quanto mais proximo de 1 melhor. 
#nunca conseguiremos explixar 100%
# consideramos de 80% a 90%

y =vendas.Vendas
y.head()
```

```python
x = vendas.Comercial
x = sm.add_constant(x)
x
```

```python
# criação do modelo 
#OSL  = Reta dos minimos quadrados

modelo = sm.OLS(y,x)
modelo
```

```python
#Verificar o quanto o modelo esta ajustad

ajuste = modelo.fit()
ajuste
ajuste.summary()

```

```python
#tirando comercial e colocano publicidade
y =vendas.Vendas
y.head()
x = vendas.Publicidade
x = sm.add_constant(x)
modelo2 = sm.OLS(y,x)
ajuste = modelo2.fit()
ajuste.summary()
```

```python
y = vendas.Vendas
x = vendas[['Publicidade','Comercial']]
x = sm.add_constant(x)
modelo2 = sm.OLS(y,x)
ajuste = modelo2.fit()
ajuste.summary()


# coefeiciente perfeito fica entre 0 e 1
###y = const + (R-squared * o que eu quero aumentar)
#y = 41+(3.5888*60)
#y
```

```python
vendas['Previsao'] = modelo2.predict(ajuste.params)
vendas
```

```python
#erro
vendas['erro'] = vendas.Vendas  - vendas.Previsao
vendas
```

```python
# o erro do modelo precisa ter media 0
vendas.erro.mean()
```
