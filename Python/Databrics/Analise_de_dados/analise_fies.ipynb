```
import pyspark.pandas as ps
```

Construir um notebook que leia o arquivo utilizando a coluna ID do estudante como entrada para o parâmetro index_col. O separador dos dados no arquivo csv é o ponto e vírgula, por isso utilize a opção sep=’;’;


```
local_arquivo = ('/FileStore/tables/relatorio_inscricao_dados_abertos_fies_22021.csv')
```

```
df = ps.read_csv(local_arquivo,sep=';',index_col='ID do estudante',encoding='ISO-8859-1')
```

Visualizar as primeiras linhas do arquivo;

```
df.head()
```

```
df.shape
```

Eliminar valores do tipo NaN da tabela;

```
df.dropna(inplace=True)
df.head()
```

```
df.shape
```

Fazer uma query para que possamos ter uma noção da proporção de homens e mulheres no dataset. Para isso você pode contar as ocorrências de valores individuais na coluna Sexo. No SQL você pode usar a função COUNT para fazer essa tarefa;

```
agrupado_sexo = ps.sql('''
SELECT Sexo , count(*) as Freq
FROM {DF}
GROUP BY Sexo
       ''',DF=df)
```

```
agrupado_sexo.plot.bar(x='Sexo',y='Freq')
```
